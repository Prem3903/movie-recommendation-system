```python
import ast

def convert(obj):
    # If it's a string, evaluate it to a Python object
    if isinstance(obj, str):
        try:
            obj = ast.literal_eval(obj)
        except Exception as e:
            print("Error converting:", obj)
            return []

    # Now obj should be a list
    if isinstance(obj, list):
        # If it's a list of dicts, extract 'name' values
        if obj and isinstance(obj[0], dict):
            return [i['name'] for i in obj if 'name' in i]
        # If it's already a list of strings, return as-is
        elif obj and isinstance(obj[0], str):
            return obj

    return []
```

```python
import ast

def convert3(obj):
    # If it's a string, evaluate it to a Python object
    if isinstance(obj, str):
        try:
            obj = ast.literal_eval(obj)
        except Exception as e:
            print("Error converting:", obj)
            return []

    # Now obj should be a list
    if isinstance(obj, list):
        # If it's a list of dicts like [{'name': 'Action'}, ...]
        if obj and isinstance(obj[0], dict):
            return [i['name'] for i in obj[:3] if 'name' in i]  # Take only first 3
        # If it's a list of strings like ['Action', 'Comedy', ...]
        elif obj and isinstance(obj[0], str):
            return obj[:3]  # Already a list of names  take first 3

    return []
```

```python
import ast

def get_directors(crew_str):
    try:
        crew_list = ast.literal_eval(crew_str) if isinstance(crew_str, str) else crew_str
        return [person['name'] for person in crew_list if person.get('job') == 'Director']
    except Exception as e:
        print("Error parsing crew data:", e)
        return []
```

```python
movies['crew'] = movies['crew'].apply(get_directors)
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Superman')
```
